```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

# Initialize the WebDriver (assuming Chrome here)
driver = webdriver.Chrome()

# Set to store unique hotel URLs
hotel_urls = set()

# Loop through the dates
for i in range(0, 30):
    # Construct the URL for each date range
    driver.get(f"https://sa.wego.com/en/hotels/searches/ruh/2024-10-{i}/2024-10-{i+1}?bow_only=true&wg_source=google&wg_medium=sem&wg_campaign=21038919720&wg_adgroup=158429811359&wg_content=691597205952&wg_term=p&wg_misc=%7Ckwd-330258448999%7C&gad_source=1&gclid=CjwKCAjwuMC2BhA7EiwAmJKRrGpP86HqkfS9czKNm6eNejZ3kLj9aoFMDlxM2M4nX4pT7AXqCijEAhoCKnAQAvD_BwE&sort=popularity&order=desc&guests=2")
    
    try:
        last_height = driver.execute_script("return document.body.scrollHeight")

        while True:
            # Scroll down to load more content
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(2)  # Adjust this time for loading more elements

            # Capture all the hotel URLs on the page
            hotels = driver.find_elements(By.CSS_SELECTOR, "a.ydPANgiIQ1BSabj12_7E[href]")

            for hotel in hotels:
                hotel_url = hotel.get_attribute('href')
                if hotel_url and hotel_url not in hotel_urls:
                    hotel_urls.add(hotel_url)
                    print(f"Saved URL: {hotel_url}")

            # Check the current height of the page after scrolling
            new_height = driver.execute_script("return document.body.scrollHeight")
            if new_height == last_height:
                break  # If the height hasn't changed, we are at the bottom of the page

            last_height = new_height

    except Exception as e:
        print(f"An error occurred while trying to fetch hotel URLs for 2024-10-{i} to 2024-10-{i+1}: {e}")

# Print the total number of unique URLs captured
print(f"Total URLs captured: {len(hotel_urls)}")

# Close the driver after all iterations
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import pandas as pd

# Initialize the WebDriver
driver = webdriver.Chrome()

# Initialize an empty DataFrame to store the hotel features
data = pd.DataFrame()

for i in range(0, 10):#
    url = lis[i]#
    # request the page 
    driver.get(url)
    

    # Wait until the "Show more" button is clickable, then click it
    try:
        show_more_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, "//*[@id='Details']/div[2]/div[2]/div[10]"))
        )
        show_more_button.click()
        
        # Wait for additional content to load after clicking "Show more"
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, "//div[@class='zUO6nZWeYmoZhNHXn4RB']"))
        )
    except Exception as e:
        print(f"No 'Show more' button or other error: {e}")

    # Extract the body and parse it
    soup = BeautifulSoup(driver.page_source, "html.parser")

    # Extract the features of the hotel
    name = soup.find('div', {'class': 'BiT_SVNSWNOYZETnrT3y'}).text if soup.find('div', {'class': 'BiT_SVNSWNOYZETnrT3y'}) else 'N/A'
    rate = soup.find('div', {'class': 'jjztdw9kzptPG79X4KX0'}).text if soup.find('div', {'class': 'jjztdw9kzptPG79X4KX0'}) else 'N/A'
    review = soup.find('div', {'class': 'dtXzSBVWMpq3Pm95Qa5Z'}).text if soup.find('div', {'class': 'dtXzSBVWMpq3Pm95Qa5Z'}) else 'N/A'


    divs = soup.find_all('div', {'class': 'zUO6nZWeYmoZhNHXn4RB'})

    # Initialize an empty dictionary to store the current hotel's features
    hotel_features = {
        'name': name,
        'rate': rate,
        'review': review
    }
    
    for div in divs:
        feature_name = div.text.strip()

        # If the feature exists in columns, add 1 to the current hotel features
        if feature_name in data.columns:
            hotel_features[feature_name] = 1
        else:
            # If the feature does not exist, create the column and add 1 for this row
            hotel_features[feature_name] = 1
            data[feature_name] = 0  # Initialize the column with 0s for all other hotels

    # Append the features for the current hotel to the DataFrame
    hotel_df = pd.DataFrame([hotel_features])

    # Concatenate the new hotel data with the existing DataFrame
    data = pd.concat([data, hotel_df], ignore_index=True)

# Close the WebDriver
driver.quit()

# Display the resulting DataFrame
data
```
